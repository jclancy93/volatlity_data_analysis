```python
import datetime

import pandas as pd
import pandas.io.data
from pandas import Series, DataFrame
pd.__version__

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rc('figure', figsize=(8, 7))
mpl.__version__
```

```python
series_list = []
securities = ['0388.HK', '000001.SS', '^GSPC', '^TWII']
for security in securities:
    s = pd.io.data.get_data_yahoo(security, start=datetime.datetime(2011, 2, 21), end=datetime.datetime(2016, 1, 29))['Adj Close']
    s.name = security # Rename series to match security name
    series_list.append(s)
```

```python
df = pd.concat(series_list, axis=1)
df.head()
```

```python
df = df.ffill()
df = df.bfill()
```

```python
rets = df.pct_change()
rets
```

```python
from arch import arch_model
import datetime as dt
import pandas.io.data as web
start=datetime.datetime(2011, 2, 21)
end=datetime.datetime(2016, 1, 29)
hongkong = web.get_data_yahoo('0388.HK', start=start, end=end)
returns = 100 * hongkong['Adj Close'].pct_change().dropna()
am3 = arch_model(returns, p=1, q=1)
res3 = am3.fit()
returns = (returns.rolling(window=7,center=False).std())
returns = returns + (pd.rolling_mean(returns, window=30, center=False)) * (12 ** 0.5)
xls_file = pd.ExcelFile('/Users/jack/Desktop/vhsi.xls')
vhsi = xls_file.parse('Sheet1')
vhsi = vhsi.set_index(pd.DatetimeIndex(vhsi['Date']))
hongkong = hongkong.merge(vhsi,how='left', left_on='Open', right_on='Close')
vseries = vhsi[vhsi.columns[1]]
vseries = vseries / (12 ** 0.5)
fig = plt.figure()
fig = res3.plot(annualize='M')
fig = returns.plot()
fig = vseries.plot()
res3.summary()
```

```python
### Calculate Standard Errors ###
vhsi['realized volatility'] = returns
vhsi['GARCH'] = res3.conditional_volatility
vhsi2 = vhsi.ix[35:]
vhsi2['Close'] = vhsi['Close'] / (12 ** 0.5)
vhsi2['GARCH'] = vhsi2['GARCH'] * (12 ** 0.5)
```

```python
vhsi2['MSE-GARCH-REAL'] = (vhsi2['GARCH']-vhsi2['realized volatility'])**2
vhsi2['MSE-GARCH-REAL'].sum()
5724.448100645928/len(vhsi2)
####### MSE = 4.6016 between Garch and Real #########
```

```python
vhsi2['MSE-VHSI-REAL'] = (vhsi2['Close']-vhsi2['realized volatility'])**2
vhsi2['MSE-VHSI-REAL'].sum()/len(vhsi2)
####### MSE = 7.713 between VHSI and Real #########
```
